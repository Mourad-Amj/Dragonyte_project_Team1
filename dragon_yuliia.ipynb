```python
# read data into a DataFrame
market = pd.read_csv('Market_Sizes.csv')
market.head(3)
```

```python
market.info()
```

```python
duplicate = market[market.duplicated ()]
print(duplicate)
```

```python
market.groupby (market.columns.tolist (), as_index= False ). size ()
```

```python
# Convert the 'date' column to datetime format
market['Year_date'] = pd.to_datetime(market['Year_date'], format='mixed', dayfirst=True)

# Change the date format to "31-12-2016"
market['Year_date'] = market['Year_date'].dt.strftime('%d-%m-%Y')
market.head(3)
```

```python
market.to_csv('Market_Sizes_clean.csv', index=False)
```
